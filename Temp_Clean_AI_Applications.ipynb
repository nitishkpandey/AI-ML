```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_org=pd.read_csv(r'/content/bank_data.csv')
print(df_org.shape)
print("Count of nan:", df_org.isna().sum())
display(df_org.head(10))
```

Task 1: Apply function on columns gender to convert all og its f to q and all of its m to 0.

Task 2: Create a new colunm called churn in which whenever the value of column retained is 0, churn should be 1 and vice versa.


```python
df = df_org.copy()

def convert(x):
  if x == 'f':
    return '1'
  else:
    return '0'

df['gender']= df["gender"].apply(convert)

display(df_org.head(10))
```

```python
def convert2(x):
  if  x== 0:
    return 1
  else:
    return 0

df["churn"] =df["retained"].apply(convert2)

display(df_org.head(10))
```

```python
df['gender'] = df['gender'].apply(lambda x: 1 if x == "f" else 0)
display(df.head(10))

df['churn'] = df['retained'].apply(lambda x: 1 if x == 0 else 0)
display(df.head(10))
```

```python
df.drop(['id','last_name','retained'], axis=1, inplace=True)
df.head()
```
