#### **Task**
Write a Python program using NLTK's WordNet to find and print a unique list of all synonyms for the word "car".

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet

synonyms = []
for syn in wordnet.synsets('car'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
```

```python
unique_synonyms = list(set(synonyms))
for syn in unique_synonyms:
    print(syn)
```
